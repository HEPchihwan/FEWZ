```python
import numpy as np 
import matplotlib.pyplot as plt

output = np.loadtxt('H_T_Histogram_Extracted.csv',delimiter=',', skiprows=1)
x = output[5:, 0]
y = output[5:, 3]
plt.plot(x, y, 'o-')
#plt.yscale('log')
plt.xlabel('bin')
plt.ylabel('kfactor')
plt.xticks(x, rotation=90)
plt.title('bin , kfactor')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

output = np.loadtxt('H_T_Histogram_Extracted.csv',delimiter=',', skiprows=1)
x = output[:, 0]
NLO = output[:, 1]
LO = output[:, 2]
plt.plot(x, LO, '--')
plt.plot(x, NLO, '--')
#plt.yscale('log')
plt.xlabel('HT bin')
plt.ylabel('cross section (Log scale)')
plt.xticks(x, rotation=90)
plt.xticks(fontsize=6)
plt.legend(['LO', 'NNLO'])
plt.title('HT bin - cross section')
plt.grid()
plt.show()
```
